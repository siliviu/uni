```python
from random import choices, sample
from math import comb, perm

IT = 100000

a = ['r']*5+['a']*3+['v']*2

cntA=0
cntAB=0
for _ in range(IT):
    e = sample(a,3)
    if 'r' in e:
        cntA+=1
        if e==['a']*3 or e==['r']*3:
            cntAB+=1

print(cntAB/cntA)
print(comb(5,3)/(comb(10,3)-comb(5,3)))

# P(B|A)=P(A ^ B)/P(A)=1/comb(5,3)
# P(A)= (comb(10,3)-comb(5,3))/comb(10,3)
# P(A^B)=comb(5,3)/comb(10,3)
```

```python
from random import randrange
from matplotlib.pyplot import bar, hist, grid, show, legend

data = [randrange(1,7) for _ in range(500)]
bin_edges = [k+0.5 for k in range(7)]
hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',
alpha = 0.5, label = 'frecvente relative')
distribution = dict([(i,1/6) for i in range(1,7)])
bar(x= distribution.keys(),height=distribution.values(), width = 0.85, color = 'red', edgecolor = 'black',
alpha= 0.6, label = 'probabilitati')
legend(loc = 'lower left')
grid()
show()
```
