```python
from scipy.stats import bernoulli, binom
from matplotlib.pyplot import bar, hist, grid, show, legend, xticks


p = 0.5
STEPS = 10
IT = 1000
N = 7

def a():
    x = bernoulli.rvs(p, size = STEPS)
    pos = 0
    for y in x:
        if y == 1:
            pos += 1
        else:
            pos -= 1
        print("Position :", pos)

def b():
    final = []
    
    for _ in range(IT):
        x = bernoulli.rvs(p, size = STEPS)
        pos = 0
        for y in x:
            if y == 1:
                pos += 1
            else:
                pos -= 1
        final.append(pos)

    # print(final)

    data = final
    bin_edges = [k+0.5 for k in range(-STEPS-1,STEPS+1)]
    hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',alpha = 0.5, label = 'frecvente relative')
    xticks(range(-STEPS,+STEPS+1))
    grid()
    show()

    # pf = 6 <-> 2nd-10=6 <-> nd = 8 bin dist
    # print(binom.pmf(8,10,0.5))
    
def c():
    final = []
    
    for _ in range(IT):
        x = bernoulli.rvs(p, size = STEPS)
        pos = 0
        for y in x:
            if y == 1:
                pos += 1
            else:
                pos -= 1
            pos = (pos + N) % N
        final.append(pos)

    # print(final)

    data = final
    bin_edges = [k + 0.5 for k in range(-1, N)]
    hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',alpha = 0.5, label = 'frecvente relative')
    xticks(range(0,N))
    grid()
    show()



a()
b()
c()
```
