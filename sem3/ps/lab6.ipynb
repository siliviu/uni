```python
from scipy.stats import norm
from numpy import mean, std, linspace
from matplotlib.pyplot import show, hist, grid, legend, xticks, plot

M = 165
SIGMA = 10
IT = 5000

data = norm.rvs(loc = M, scale = SIGMA, size = IT)

hist(data, bins = 16,range = (130, 210), density = True, rwidth = 0.9, color = 'green', edgecolor = 'black', alpha = 0.5, label = 'frecvente relative')
x = linspace(130, 210, 1000)
y = norm.pdf(x,loc=M,scale=SIGMA)
xticks(range(130,215,5))
plot(x,y,'r-', label='functie densitate')
legend()
show()

print('mean =',mean(data), ',std deviation =',std(data), ',%[160, 170] =', mean((data >= 160) & (data <= 170)), 'vs', norm.cdf(170,loc=M,scale=SIGMA)-norm.cdf(160,loc=M,scale=SIGMA))
```

```python
IT = 5000
LAMBDA=1/5

from scipy.stats import expon, uniform
from numpy import mean, std, multiply

data = []

for _ in range(IT):
    if uniform.rvs()<0.4:
        data.append(expon.rvs(scale=1/LAMBDA))
    else:
        data.append(uniform.rvs(loc=4,scale=2))

print(mean(data), std(data))

print(mean([x < 5 for x in data]))

print(0.4 * expon.cdf(5, scale = 1 / LAMBDA) + 0.6 * uniform.cdf(5, loc = 4, scale = 2))
```

```python
from scipy.stats import uniform
from numpy import exp, mean
from scipy.integrate import quad

A = -1
B = 3
IT = 1000000

val = uniform.rvs(loc=A,scale=B-A,size=IT)
g = lambda x:exp(-x**2)

print(mean(g(val)) * (B - A))
print(quad(g, A, B)[0])
```
